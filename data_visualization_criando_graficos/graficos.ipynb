```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Arthur Lindolpho\Documents\data_science_Alura\data_visualization_criando_graficos\imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
df_brasil = pd.DataFrame(brasil_dict)
```

```python
df_brasil
```

```python
df_brasil.columns
```

```python
plt.figure(figsize=(14,8))
plt.plot(df_brasil['ano'], df_brasil['imigrantes'], marker = 'x')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.xlabel('Anos')
plt.ylabel('Número de imigrantes')
plt.title('Migração do Brasil para o Canadá')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(df_brasil['ano'], df_brasil['imigrantes'], lw = 3, color = 'g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel('Ano', fontsize = 14)
ax.set_title('Imigração do Brasil para o Canadá\n 1980 a 2013', fontsize = 18, loc = 'left')
ax.set_ylabel('Número de imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
#plt.grid(linestyle = '--')
fig.savefig('imigracao_brasil_canada.png', transparent=False, dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []
for pais in america_sul.index:
    if pais == 'Brasil':
        cores.append('green')
    else: 
        cores.append('silver')
```

```python
fig, ax = plt.subplots(figsize = (12,8))
barras = ax.barh(america_sul.index, america_sul['Total'], color = cores)
ax.bar_label(barras, america_sul['Total'], fontweight = 'bold', fontsize = 8)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes \n para o Canadá de 1980 a 2013', loc = 'left', fontsize = 18)
ax.set_xlabel('Número de imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.set_xticks([])
ax.tick_params(axis = 'y', length = 0)
ax.set_frame_on(False)
fig.savefig('imigracao_america_sul.png', transparent=False, dpi = 300, bbox_inches='tight')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2)
fig.set_size_inches(15,5)
axs[0].plot(df_brasil['ano'], df_brasil['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_title('Imigração do Brasil para o Canadá\n 1980 a 2013')
axs[0].set_ylabel('Número de imigrantes')
axs[0].grid()

axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá \n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2)
fig.set_size_inches(16,8)
fig.suptitle('Imigração dos 4 maiores países da Amarérica do Sul para o Canadá \n 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0,0].set_title('Brasil')
axs[0,0].grid()

axs[0,1].plot(df.loc['Argentina', anos])
axs[0,1].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0,1].set_title('Argentina')
axs[0,1].grid()

axs[1, 0].plot(df.loc['Colômbia', anos])
axs[1, 0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[1, 0].set_title('Colômbia')
axs[1, 0].grid()

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[1, 1].set_title('Peru')
axs[1, 1].grid()

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
    ax.set_ylim(0 , 7000)

plt.tight_layout()
plt.show()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = top_10.index, y = top_10['Total'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
sns.barplot(y = top_10.index, x = top_10['Total'], palette='viridis', hue=top_10.index)
plt.title('Países com maior imigração para o Canadá \n 1980 a 2013', loc = 'left')
plt.xlabel('Número de imigrantes')
plt.ylabel('')
plt.show()
```

```python
sns.set_theme(style='ticks')
fig, ax = plt.subplots(figsize = (10, 8))
sns.barplot(y = top_10.index, x = top_10['Total'], palette='viridis', hue=top_10.index, ax = ax)
for container in ax.containers:
    ax.bar_label(container, padding=-80, color = 'white', weight = 'bold')
plt.title('Países com maior imigração para o Canadá \n 1980 a 2013', loc = 'left')
plt.xlabel('Número de imigrantes')
plt.ylabel('')
plt.box(False)
plt.tick_params(axis = 'both', length = 0)
plt.xticks([])
plt.show()
```
