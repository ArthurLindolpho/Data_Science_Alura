## Analisando a distribuição dos dados

```python
def largura_bin(df, coluna):
    Q75, Q25 = np.percentile(df[coluna], [75, 25])
    IQR = Q75 - Q25
    largura_bin = 2 * IQR * np.power(len(df[coluna]), -1/3)
    return largura_bin
```

```python
largura_bon_at = largura_bin(dados, 'arrival_time')
```

```python
sns.histplot(data = dados, x = 'arrival_time', kde=True, binwidth = largura_bon_at)
```

```python
largura_bin_dt = largura_bin(dados, 'departure_time')
```

```python
sns.histplot(data = dados, x = 'arrival_time', kde=True, binwidth = largura_bin_dt)
```

```python
mean_delay = dados['delay'].mean()
median_delay = dados['delay'].median()

fig, axs = plt.subplots(1, 2, figsize = (9, 4))
sns.boxplot(data = dados, y ='delay', ax = axs[0])
axs[0].set_title('Boxplot')
axs[0].axhline(y = mean_delay, color = 'r', linestyle = '--', label = 'Média')
axs[0].legend()

largura_bin_delay = largura_bin(dados, 'delay')
sns.histplot(data = dados, x = 'delay', ax = axs[1], kde = True, binwidth=largura_bin_delay)
plt.ylabel('Número de voos')
axs[1].set_title('Histograma')

axs[1].axvline(x = mean_delay, color = 'r', linestyle = '--', label = 'Média')
axs[1].axvline(x = median_delay, color = 'y', linestyle = '--', label = 'Mediana')
axs[1].legend()

plt.tight_layout()
plt.show()
```

## Seleção e validação do modelo de Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

### Treinamento do Dummy Regressor

Este é nosso modelo base, os próximos modelos devem superar as métricas deste modelo

```python
x = df.drop(columns=['delay'])
y = df['delay']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dummy = DummyRegressor()
dummy.fit(x_train, y_train)
```

```python
y_pred = dummy.predict(x_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f'''
R²: {r2 :.2f}
MAE: {mae :.2f}
RMSE: {rmse :.2f}
''')
```

### Treinando o modelo Random Forest

```python
model_RF = RandomForestRegressor(max_depth= 10, random_state=42, n_estimators=150)
model_RF.fit(x_train, y_train)
y_pred_RF = model_RF.predict(x_test)

r2 = r2_score(y_test, y_pred_RF)
mae = mean_absolute_error(y_test, y_pred_RF)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_RF))

print(f'''
R²: {r2 :.2f}
MAE: {mae :.2f}
RMSE: {rmse :.2f}
''')
```

```python
from yellowbrick.regressor import prediction_error

visualizar = prediction_error(model_RF, x_train, y_train, x_test, y_test)
```

```python
from yellowbrick.regressor import residuals_plot

viz = residuals_plot(model_RF, x_train, y_train, x_test, y_test)
```

## Validação Cruzada

```python
from sklearn.model_selection import KFold, cross_validate
```

```python
scoring = {
    'mae': 'neg_mean_absolute_error',
    'rmse': 'neg_root_mean_squared_error',
    'r2': 'r2'
}

cv = KFold(n_splits= 5, shuffle=True, random_state=42)

cv_results = cross_validate(model_RF, x_train, y_train, cv = cv, scoring= scoring)
cv_results
```

```python
for metric in scoring.keys():
    scores = cv_results[f'test_{metric}']
    mean_score = '{:.3f}'.format(scores.mean())
    std_score = '{:.3f}'.format(scores.std())

    scores_list = [f'{val:.3}' for val in scores]
    print(f'{metric.upper()} Scores: {scores_list}')
    print(f'{metric.upper()} Média: {mean_score}, Std: {std_score}')
    print('-='*40)
```

## Otimização de hiperparâmetros

### Seleção de recursos

```python
from yellowbrick.model_selection import FeatureImportances

viz = FeatureImportances(model_RF)
viz.fit(x_train, y_train)
viz.show()
```

```python
importances = model_RF.feature_importances_
feature_importances = pd.DataFrame({'Feature': x.columns, 'Importances': importances})
feature_importances.sort_values('Importances', ascending=False)
```

```python
def metricas(y_test, y_pred):
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    metrics = {
        'RMSE': round(rmse, 4),
        'MAE': round(mae, 4),
        'R2': round(r2, 4)
    }

    return metrics
```

```python
results_df = pd.DataFrame(index = ['RMSE', 'MAE', 'R2'])
model_selected_feature = RandomForestRegressor(random_state=42, max_depth=5)

for count in [1, 5, 10, 15, 20, 25, 30]:
    selected_features = feature_importances['Feature'].values[:count]

    X_train_selected = x_train[selected_features]
    X_test_selected = x_test[selected_features] 

    model_selected_feature.fit(X_train_selected, y_train)

    y_pred = model_selected_feature.predict(X_test_selected)

    result_metricas = metricas(y_test, y_pred)

    results_df[count] = list(result_metricas.values())

results_df
```

```python
results_df = pd.DataFrame(index = ['RMSE', 'MAE', 'R2'])
model_selected_feature = RandomForestRegressor(random_state=42, max_depth=5)

for count in range(10,16):
    selected_features = feature_importances['Feature'].values[:count]

    X_train_selected = x_train[selected_features]
    X_test_selected = x_test[selected_features] 

    model_selected_feature.fit(X_train_selected, y_train)

    y_pred = model_selected_feature.predict(X_test_selected)

    result_metricas = metricas(y_test, y_pred)

    results_df[count] = list(result_metricas.values())

results_df
```

## Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [5, 10, 15],
    'min_samples_leaf': [1,2,3],
    'min_samples_split': [2,4,6],
    'n_estimators': [50, 100, 150, 200],
}
```

```python
cv = KFold(n_splits= 5, shuffle=True, random_state=42)
```

```python
model_grid = GridSearchCV(RandomForestRegressor(random_state=42), param_grid = param_grid, scoring= 'r2', cv = cv, n_jobs=1)
model_grid.fit(x_train, y_train)
```

```python
model_grid.best_params_
```

```python
y_pred_model_grid = model_grid.predict(x_test)
metricas_model_grid = metricas(y_test, y_pred_model_grid)
metricas_model_grid
```
