```python
colunas = df.columns
colunas
```

```python
one_hot = make_column_transformer((
    OneHotEncoder(drop = 'if_binary'),
    ['estado_civil', 'escolaridade', 'inadimplencia', 'fez_emprestimo', 'aderencia_investimento']
),
    remainder= 'passthrough',
    sparse_threshold=0)
```

```python
df = one_hot.fit_transform(df)
```

```python
one_hot.get_feature_names_out(colunas)
```

```python
df_cod = pd.DataFrame(df, columns= one_hot.get_feature_names_out(colunas))
```

```python
df_cod.columns
```

```python
df_cod['onehotencoder__aderencia_investimento_sim'].value_counts()
```

```python
# Variável alvo
y = df_cod['onehotencoder__aderencia_investimento_sim']

# Variáveis explicativas
X = df_cod.drop(columns='onehotencoder__aderencia_investimento_sim')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify= y, test_size= 0.2, random_state=5)
```

```python
dummy = DummyClassifier()
dummy.fit(X_train, y_train)
dummy.score(X_test, y_test)
```
